```python
from mtcnn import MTCNN
import cv2
import numpy as np
import os

outer_path = '/root/FACIAL/video_preprocess/train1_image/'
filelist = os.listdir(outer_path)
detector = MTCNN()
n=1

for item in filelist:
    if n < 1:
      n += 1
      continue
    src = os.path.join(os.path.abspath(outer_path),str(n).zfill(6)+".jpg")
    print(src)
    input_img = cv2.imread(src)
    img = cv2.cvtColor(input_img, cv2.COLOR_BGR2RGB)

    detector.detect_faces(img)
    face_arr = detector.detect_faces(img)
    face=face_arr[0]['keypoints']
    print(face,end='\n')
    print(str(n)+'\n')

    f=open("/root/autodl-fs/detections/"+str(n).zfill(6)+".txt","w")
    f.write(str(face['left_eye'][0])+'\t'+str(face['left_eye'][1])+'\n')
    f.write(str(face['right_eye'][0])+'\t'+str(face['right_eye'][1])+'\n')
    f.write(str(face['nose'][0])+'\t'+str(face['nose'][1])+'\n')
    f.write(str(face['mouth_left'][0])+'\t'+str(face['mouth_left'][1])+'\n')
    f.write(str(face['mouth_right'][0])+'\t'+str(face['mouth_right'][1]))
    f.close()
    n=n+1

print("successful")
```
